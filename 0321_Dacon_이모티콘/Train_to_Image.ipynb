```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 불러오기
df = pd.read_csv("train.csv")

# 2. X, y 분리
X = df.iloc[:, 2:]  # 이미지 벡터 (1024차원)
y = df['label']     # 이모티콘 라벨

# 3. 시각화: 앞 100개 (10x10 격자)
fig, axes = plt.subplots(16, 16, figsize=(12, 12))
fig.suptitle("100", fontsize=16)

for i, ax in enumerate(axes.flat):
    image = X.iloc[i].values.reshape(32, 32)
    label = y.iloc[i]
    ax.imshow(image, cmap='gray')
    ax.set_title(str(label), fontsize=8)
    ax.axis('off')

plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```
