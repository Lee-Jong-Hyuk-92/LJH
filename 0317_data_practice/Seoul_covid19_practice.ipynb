```python
covid_per_day = df_covid["확진일"].value_counts() # covid_per_day에 확진일 같은 값들을 카운트해서 저장
covid_per_day
```

```python
covid_per_day.sort_values(ascending=False).head() # 내림차순으로 정렬, 상위 5개만
```

```python
covid_per_day[covid_per_day == covid_per_day.max()] # covid_per_day의 값 중에서 covid_per_day의 max 값과 같은 covid_per_day를 출력
```

```python
covid_per_day = covid_per_day.sort_index() # covid_per_day를 index 기준(day)으로 정렬
covid_per_day
```

```python
plt.figure(figsize=(15, 10)) # 사이즈 15인치, 10인치로 설정
plt.plot(covid_per_day)

plt.title("COVID-19 infected per day")
plt.xlabel("Date")
plt.ylabel("Number of infected people")
plt.xticks(covid_per_day.index[::60])

plt.show()
```

```python
covid_per_area = df_covid["지역"].value_counts() # 이번에는 지역 열로 묶어서 어디가 가장 많이 나왔는지 보자
covid_per_area
```

```python
colors = pd.Series("gray", index=covid_per_area.index) # colors 회색으로 설정
colors["관악구"] = "red" # colors 중에서 '관악구'만 빨간색 설정

plt.bar(covid_per_area.index, covid_per_area, color=colors) # 지역마다 감염자 수를 bar 차트로 플랏, 회색 기본, 관악구는 빨강
```

```python
fig = plt.figure(figsize=(20, 12)) # 가로 20인치, 세로 12인치

colors = pd.Series("gray", index=covid_per_area.index) # 색은 동일하게 그레이 기본
colors["관악구"] = "red" # 관악구만 빨강

plt.bar(covid_per_area.index, covid_per_area, color=colors)

plt.title("Covid-19 Infected per Area", fontsize=24)
plt.xlabel("Area", fontsize=14)
plt.ylabel("Number of Infected People", fontsize=14)

plt.show()
```
