```python
# 순열, 퍼뮤테이션, P, n개중 r개를 순서생각하면서  뽑기, nPr = n!/(n-r)! 7P4=7*6*5*4, 7부터 1씩 줄여가며 4개 곱한다
from itertools import permutations

p1 = list(permutations([1,2,3,4,5,6],3))

print(p1, '\n',len(p1))
```

```python
# 조합, 콤비네이션, C, n개중 r개를 순서상관없이 뽑기, nCr = n!/r!(n-r)!
from itertools import combinations

c1 = list(combinations([1,2,3,4,5,6],3))
print(c1,'\n',len(c1))
```

```python
# 중복순열, Product, n개중 r개를 중복 허용하면서 순서생각하면서 뽑기기, nRPr = n^r
from itertools import product

pi1 = list(product([1,2,3,4,5,6],repeat=3))
print(pi1,'\n',len(pi1))
```

```python
# 중복 조합,  Combinations_with_replacement, n개중 r개를 중복허용, 순서상관없이 뽑기, nRCr = n+r-1Cr = (n+r-1)!/r!(n-1)!
from itertools import combinations_with_replacement

h1 = list(combinations_with_replacement([1,2,3,4,5,6],3))
print(h1,'\n',len(h1))
```
